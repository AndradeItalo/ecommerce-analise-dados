```python
import pandas as pd
```

ler todos os arquivos, em formato de texto

```python
customers = pd.read_csv("../data/raw/customers.csv", dtype=str)
products = pd.read_csv("../data/raw/products.csv", dtype=str)
orders = pd.read_csv("../data/raw/orders.csv", dtype=str)
order_items = pd.read_csv("../data/raw/order_items.csv", dtype=str)
reviews = pd.read_csv("../data/raw/reviews.csv", dtype=str)
```

```python
tables = {
    "customers": customers,
    "products": products,
    "orders": orders,
    "order_items": order_items,
    "reviews": reviews,
}
```

IDS DUPLICADOS
 se df.duplicated().sum() for igual a df['customer_id'].duplicated().sum(), os duplicados são cópias exatas
 mas se df['customer_id'].duplicated().sum() for maior, significa que existe o mesmo customer_id aparecendo com dados diferentes (ex.: email diferente em cada linha)

```python
id_columns = {
    "customers": ["customer_id"],
    "products": ["product_id"],
    "orders": ["order_id", "customer_id"],
    "order_items": ["order_item_id", "order_id", "product_id"],
    "reviews": ["review_id", "product_id", "customer_id"]
}

for table_name, cols in id_columns.items():
    print("="*60)
    print(table_name)

    df = tables[table_name]

    for col in cols:
        print(col, "->", df[col].duplicated().sum())
    
```

VER colunas categóricas

```python
categorical_columns  = {
    "customers": ["state", "customer_segment"],
    "products": ["brand","category"],
    "orders": ["status", "payment_method", "shipping_state"]
}

for table_name, cols in categorical_columns.items():

    df = tables[table_name]

    for col in cols:
        print("="*60)
        print(table_name, "-", col)

        print(df[col].value_counts(dropna=False))
```

ver orders com costumer id que nao tem na tabela de costumers
ver orders items id com products ids que nao existem na tabela de products

```python
customers_ids = set(customers["customer_id"])
orders_ids = set(orders["customer_id"])

missing = orders_ids - customers_ids

print("Clientes inexistentes:", len(missing))

products_ids = set(products["product_id"])
items_ids = set(order_items["product_id"])

missing = items_ids - products_ids

print("Produtos inexistentes:", len(missing))
```

## RESUMO FINAL DOS ACHADOS — mapa de sujeira do dataset

### customers (3.060 linhas)
- **PK duplicada:** 60 `customer_id` repetidos (1 linha 100% idêntica; os demais têm o mesmo ID com dados diferentes — precisa decidir qual registro manter na Silver).
- **Nulos relevantes:** `customer_segment` (487), `phone` (256), `cpf` (300), `signup_date` (159), `birth_date` (148), `email` (145).
- **Datas (`birth_date`, `signup_date`):** pelo menos 5 formatos misturados — ISO com hora (`1991-10-19T00:00:00`), `dd/mm/yyyy` com e sem hora `00:00`, `dd-mm-yyyy`, `yyyy/mm/dd`, `mm/dd/yyyy`, e o sentinela inválido `0000-00-00`.
- **Categóricas sujas:** `state` com maiúsc/minúsc misturado (`df`/`DF`, `sp`/`SP`...); `customer_segment` com `NOVO`/`novo` duplicando o mesmo valor.
- **Outros:** `first_name`/`last_name` mal segmentados — títulos como "Sr.", "Srta.", "Dr." aparecem em `first_name` no lugar do nome; `phone` e `cpf` com e sem máscara/formatação.

### products (606 linhas)
- **PK duplicada:** 6 `product_id` repetidos.
- **Nulos relevantes:** `stock_quantity` (135), `created_at` (37), `price` (22).
- **Sentinela inválido:** `stock_quantity` com valores negativos (ex.: `-1`, `-5`) — estoque não pode ser negativo, é um marcador de "sem dado" disfarçado de número.
- **Dinheiro (`price`):** 3 formatos — `1428.71` (ponto decimal), `"8983,68"` (vírgula BR), `"R$ 1.590,09"` (moeda BR completa com separador de milhar).
- **Datas (`created_at`):** mesma mistura de formatos de `customers`, incluindo `0000-00-00`.
- **Encoding/categoria (`category`):** a mesma categoria escrita em até 5 grafias diferentes por causa de caixa E mojibake de encoding — ex. "eletrônicos": `Eletrônicos`, `ELETRONICOS`, `eletronicos`, `Eletronicos`, `EletrÃ´nicos` (UTF-8 lido como Latin-1). Mesmo padrão em "informática", "decoração" e "móveis".

### orders (15.088 linhas)
- **PK duplicada:** 88 `order_id` repetidos (linhas 100% idênticas).
- **FK órfã:** 210 `customer_id` que não existem em `customers` — pedidos "fantasma" sem cliente.
- **Nulos relevantes:** `payment_method` (1.969), `total_amount` (1.298), `order_date` (787).
- **Categórica mais suja do dataset (`status`):** ~24 variações do mesmo conjunto de 5 estados reais, misturando português/inglês, maiúsc/minúsc e até espaço em branco à direita (`entregue` aparece 2x com contagens diferentes por causa disso). Precisa de um de-para explícito PT↔EN + normalização de caixa/espaço.
- **`payment_method`:** 3 grafias para "cartão de crédito" (`cartao credito`, `CARTAO_CREDITO`, `cartao_credito`) e caixa mista em `Pix`/`pix`.
- **Datas (`order_date`):** mesmos formatos misturados, incluindo timestamp completo (`2025-08-06T18:40:21`) e o formato americano inequívoco `10/25/2025` (dia 25 prova que não pode ser `dd/mm`).
- **Dinheiro (`total_amount`):** mesmos 3 formatos de `price`.

### order_items (26.191 linhas)
- **PK limpa:** `order_item_id` sem nenhum duplicado — a única chave primária 100% íntegra do dataset.
- **FK órfã:** 138 `product_id` (na faixa 601–800) que não existem em `products` (que só vai até 600).
- **Nulos relevantes:** `unit_price` (1.285), `discount_pct` (1.320).
- **Dinheiro (`unit_price`):** mesmos 3 formatos das outras colunas monetárias.
- **Nota conceitual:** `order_id` e `product_id` aqui são chaves estrangeiras — números altos de "duplicados" nessas colunas (11.191 e 25.453) são esperados (muitos itens por pedido, muitos itens do mesmo produto) e não representam sujeira.

### reviews (8.080 linhas)
- **PK duplicada:** 80 `review_id` repetidos (linhas 100% idênticas).
- **FKs íntegras:** nenhum `product_id` ou `customer_id` órfão.
- **Nulos relevantes:** `review_text` (1.630), `rating` (400), `review_date` (391).
- **Datas (`review_date`):** mesma mistura de formatos + timestamp com hora (`08/01/2025 15:21`) + sentinela `0000-00-00`.
- **Outros:** vários `review_text` repetidos (mesmo texto genérico em várias linhas) — não é erro de formato, mas indica que o texto pode ter baixo valor analítico se usado sem cautela.

### Problemas transversais (aparecem em várias tabelas — viram funções reutilizáveis na Silver)
1. **Datas em 5+ formatos** por coluna, todas com o sentinela `0000-00-00` como "nulo disfarçado".
2. **Dinheiro em 3 formatos** (`1234.56` / `"1234,56"` / `"R$ 1.234,56"`) em `price`, `total_amount`, `unit_price`.
3. **Categóricas com variação de caixa** em quase toda coluna categórica (`state`, `status`, `payment_method`, `customer_segment`, `category`).
4. **Mojibake de encoding** isolado em `products.category`.
5. **PKs duplicadas** em 4 das 5 tabelas (só `order_items` está limpa).
6. **Integridade referencial quebrada** em 2 relacionamentos: `orders→customers` (210 órfãos) e `order_items→products` (138 órfãos); `reviews` está 100% íntegra.
7. **Sentinelas de nulo disfarçados de número válido:** `stock_quantity` negativo.

Este mapa é o backlog de decisões da camada **Silver**: cada item acima vira uma regra de limpeza explícita e testável.
