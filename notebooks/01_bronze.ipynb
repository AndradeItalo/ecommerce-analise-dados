```python
import pandas as pd
```

```python
customers = pd.read_csv("../data/raw/customers.csv", dtype=str)
products = pd.read_csv("../data/raw/products.csv", dtype=str)
orders = pd.read_csv("../data/raw/orders.csv", dtype=str)
order_items = pd.read_csv("../data/raw/order_items.csv", dtype=str)
reviews = pd.read_csv("../data/raw/reviews.csv", dtype=str)
```

```python
tables = {
    "customers": customers,
    "products": products,
    "orders": orders,
    "order_items": order_items,
    "reviews": reviews,
}

for name, df in tables.items():
    df["_ingested_at"] = pd.Timestamp.now()
    df["_source_file"] = f"{name}.csv"
```

```python
for name, df in tables.items():
    df.to_parquet(
        f"../data/bronze/{name}.parquet",
        index=False
    )
```
